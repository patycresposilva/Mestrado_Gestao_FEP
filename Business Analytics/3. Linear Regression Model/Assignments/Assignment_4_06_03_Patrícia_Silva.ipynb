# REGRESSION PROBLEM
Dataset: White wine

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/winequality-white.csv', sep=';')
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df.drop('quality', axis=1) #axis=1 to instruct python to consider 'quality' as a column
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=45)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.pipeline import Pipeline
```

```python
scaler = Pipeline([
    ('scale', StandardScaler())
])
```

```python
from sklearn.compose import ColumnTransformer

preprocesser = ColumnTransformer([
    ('scale2', scaler, x.columns.to_list())#'scale2' is the name of the step scaler  inside the 'pipeline' ColumnTranformer. #This name can be whatever name we like
],
  remainder = 'passthrough' #this is to pass to transformed columns to the model
)
```

```python
pipe = Pipeline([
        ('pre', preprocesser),
        ('ln', LinearRegression())
    ])

pipe.fit(x_train, y_train)
```

```python
y_preds = pipe.predict(x_train)

y_preds
```
